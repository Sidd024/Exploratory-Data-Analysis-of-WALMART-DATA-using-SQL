```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
import seaborn as sns
walmart_dataset = pd.read_csv("Walmart-Data.csv")
```

```python
#Country Wise Sum Of Sales Trend
countrywise_sales = walmart_dataset.groupby(['Country'])['Sum of Sales'].sum().reset_index().sort_values(by='Sum of Sales', ascending=False).head(6)
plt.figure(figsize=(16,5))
bars = plt.bar(countrywise_sales['Country'],countrywise_sales['Sum of Sales'])
plt.title("Country Wise Sum Of Sales Trend")
plt.xlabel("List Of Country Names")
plt.ylabel("List Of Sum Of Sales")
plt.bar_label(bars, fmt='%.0f')
plt.xticks(rotation=45)
plt.show()
```

```python
#Year Wise Sum Of Profit Trend
yearwise_profit_trend = walmart_dataset.groupby(['Country'])['Sum of Profit'].sum().reset_index().sort_values(by='Sum of Profit',ascending=False).head(6)
plt.figure(figsize=(16,5))
plt.plot(yearwise_profit_trend['Country'],yearwise_profit_trend['Sum of Profit'],marker='o')
#Value labels
for x, y in zip(
    yearwise_profit_trend['Country'], 
    yearwise_profit_trend['Sum of Profit']
):
  plt.text(x, y, f'{y:.0f}', ha='center', va='bottom')
plt.xlabel("List Of Countries Name")
plt.ylabel("Range Of Sum Of Profits")
plt.title("Top 6 Countries  Sum of Profit")
plt.show()
```

```python
#Sub-Category Wise Sum Of Profit
regionwise_profit = walmart_dataset.groupby(['Sub-Category'])['Sum of Profit'].sum().reset_index().sort_values(by='Sum of Profit',ascending=False).head(7)
plt.figure(figsize=(16,5))
plt.title("Sub-Category Wise Sum Of Profit")
plt.pie(list(regionwise_profit['Sum of Profit']),labels = list(regionwise_profit['Sub-Category']),autopct='%1.1f%%')
plt.show()
```
